## Download and Install SQLite3
SQLite is available as precompiled binaries from the SQLite __[download page](https://www.sqlite.org/download.html)__.  If you're installing onto Windows, be sure to grab the sqlite-tools command-line tools package available on the same page.  The instructions here are for a Windows 10 install.

#### I. Start Windows PowerShell in Administrator mode and extract the downloaded zip files into a temporary folder.

```
> cd c:\Users\randre\Downloads
> mkdir sqlite3
> Expand-Archive -Path .\sqlite-tools-win32-x86-3220000.zip -DestinationPath .\sqlite3\
> Expand-Archive -Path .\sqlite-dll-win64-x64-3220000.zip -DestinationPath .\sqlite3\

> dir .\sqlite3\

Mode                LastWriteTime         Length Name
----                -------------         ------ ----
d-----        1/23/2018   4:24 PM                sqlite-tools-win32-x86-3220000
-a----        1/22/2018   1:50 PM           5506 sqlite3.def
-a----        1/22/2018   1:50 PM        1217536 sqlite3.dll
```

(**NOTE** that in addition to the .def and .dll files, there is a 'sqlite-tools' folder present which contains .exe files)

```
> dir .\sqlite-tools-win32-x86-3220000\

Mode                LastWriteTime         Length Name
----                -------------         ------ ----
-a----        1/23/2018   1:58 AM         454656 sqldiff.exe
-a----        1/23/2018   1:59 AM         855552 sqlite3.exe
-a----        1/23/2018   1:59 AM        1966080 sqlite3_analyzer.exe
```

#### II. Copy ALL of the .def, .dll and .exe files into C:\Windows\system32 (or another folder that is in your PATH). I find the easiest way to do this is to copy the .exe files "up" one level out of the sqlite-tools folder, delete the sqlite-tools folder, then copy everything at once over to where you want it. 

Starting from inside the sqlite-tools folder:

```
> copy *.exe ..\
> cd ..\
> rmdir -R .\sqlite-tools-win32-x86-3220000\
> dir

Mode                LastWriteTime         Length Name
----                -------------         ------ ----
-a----        1/23/2018   1:58 AM         454656 sqldiff.exe
-a----        1/22/2018   1:50 PM           5506 sqlite3.def
-a----        1/22/2018   1:50 PM        1217536 sqlite3.dll
-a----        1/23/2018   1:59 AM         855552 sqlite3.exe
-a----        1/23/2018   1:59 AM        1966080 sqlite3_analyzer.exe

> copy * C:\Windows\System32\
```

#### III. You can check that it worked by executing "sqlite3.exe".  You should see something like this:

```
> sqlite3.exe
SQLite version 3.22.0 2018-01-22 18:45:57
Enter ".help" for usage hints.
Connected to a transient in-memory database.
Use ".open FILENAME" to reopen on a persistent database.
```
You can exit from the DB connection using the .quit command.
```
sqlite> .quit
```
