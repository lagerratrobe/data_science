```python
import pandas as pd
from vega3 import VegaLite
```

```python
df = pd.read_json('cars.json')
df.head()
```

```python
VegaLite({
  "mark": "point",
  "encoding": {
    "y": {"type": "quantitative","field": "Acceleration"},
    "x": {"type": "quantitative","field": "Horsepower"}
  }
}, df)
```

```python
map_spec = {
  "mark": "geoshape",
  "width": 600,
  "height": 400,
  "projection": {
    "type": "albersUsa"
  },
  "data": {
        "url": "us-10m.json",
        "format": {
              "type": "topojson",
              "feature": "states"
                }
          }
}

VegaLite(map_spec)
```
