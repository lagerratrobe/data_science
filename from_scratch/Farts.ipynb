Some experiments with shaping and manipulating data in CSV files.  Here's some data that looks similar to some of our names data.

NOTE: the ```%%file``` magic needs to be in its own cell to work properly.

````python
%%file name_data.csv
id|name|locale|type
60940|California|en_us|canonical
60940|Kalifornie|de_de|canonical
60940|캘리포니아 주|ko_kr|canonical
60940|CA|none|synonym
119370|Mexico|en_us|canonical
119370|Mexique|fr_fr|canonical
119370|Mexiko|de_de|canonical
119370|멕시코|ko_kr|canonical
119370|MX|none|synonym
````

````python
import pandas as pd

# READ A CSV FILE INTO A PANDAS DATAFRAME
names_frame = pd.read_csv('./name_data.csv', sep='|')
names_frame
````

````python
names_frame.set_index(['id', 'type'])
````

````python
foo = names_frame.set_index(['id', 'type'])
pivoted_foo = foo.pivot(columns='locale')
pivoted_foo
````

````python
print pivoted_foo.loc[60940, 'canonical'].loc['name', 'ko_kr']
````
